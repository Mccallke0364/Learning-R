### 2. Make a scatterplot of `hwy` vs. `displ` using the mpg data frame. Next, map a third, numerical variable to `color`, then `size`, then both `color` and `size`, then `shape`. How do these aesthetics behave differently for categorical vs. numerical variables?


        Hwy and displ are both numerical data and they are able to be graphed in data points to show potential correlations. When using a numerical variable for color or size, the color and size can change as specifically as the numeric variable but when using shape it is more benefitial to use categorical data since there can be one shape for each category, where as with numerical it would be uninformative to have different shapes for each number since it won't indicate how close in number they are unless they are the exact same number. 

````r
ggplot(mpg, aes(x=displ, y=hwy))+
geom_point()
ggplot(mpg, aes(x=displ, y=hwy, color=cyl))+
geom_point()
ggplot(mpg, aes(x=displ, y=hwy, size=cyl))+
geom_point()
ggplot(mpg, aes(x=displ, y=hwy, color=cyl, size=cyl))+
geom_point()
ggplot(mpg, aes(x=displ, y=hwy, size=cty, color=cty))+
geom_point()
ggplot(mpg, aes(x=displ, y=hwy))+
geom_point(aes(shape=cyl))
ggplot(mpg, aes(x=displ, y=hwy))+
geom_point(aes(shape=model))
````

### 3. In the scatterplot of `hwy` vs. `displ`, what happens if you map a third variable to `linewidth`?

````r
ggplot(mpg, aes(x=displ, y=hwy, linewidth=cyl))+
geom_point()
ggplot(mpg, aes(x=displ, y=hwy))+
geom_point(aes(linewidth=cyl))
````

### 4. What happens if you map the same variable to multiple aesthetics?


        When you map the same variable to multiple aesthetics, it will help show the correlation even more. However, if you map the same categorical data to multiple aesthetics, it will also group the aesthetic of each category into one for the key so each category has a unique color size and shape. 

````r
ggplot(mpg, aes(x=displ, y=hwy, color= drv, size=drv, shape=drv))+
geom_point()
ggplot(mpg, aes(x=displ, y=hwy, color= cyl, size=cyl))+
geom_point()
ggplot(mpg, aes(x=displ, y=hwy, color= model, size=model))+
geom_point()
````

### 2. What do you need to change in the code above to save the plot as a PDF instead of a PNG? How could you find out what types of image files would work in `ggsave()`? 




Since the default "device" for ggsave is NULL, which guesses the device based on the filename extension, changing the .png to .pdf will automatically change the file type of the saved plot to a pdf. This can be determined by looking at the [documentation for ggsave()](https://ggplot2.tidyverse.org/reference/ggsave.html)
> Can either be a device function (e.g. png), or one of "eps", "ps", "tex" (pictex), "pdf", "jpeg", "tiff", "png", "bmp", "svg" or "wmf" (windows only). If NULL (default), the device is guessed based on the filename extension.

````r
ggplot(mpg, aes(x = class)) +
  geom_bar()
ggplot(mpg, aes(x = cty, y = hwy)) +
  geom_point()
ggsave("mpg-plot.pdf")
````
