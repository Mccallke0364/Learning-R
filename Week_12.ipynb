```r
library(palmerpenguins)
library(ggplot2)
library(tidyverse)
library(dplyr)
```

```r
pca_data_raw<- penguins|>
    na.omit()|>
    mutate(across(where(is.numeric), as.integer))|>
    select(where(is.numeric))|>
    prcomp(scale. = TRUE)

pca_data <- as.data.frame(pca_data_raw$x)
# pca_data
```

```r
penguins_clean <- na.omit(penguins)
pca_data$species <- penguins_clean$species
pca_data$sex <- penguins_clean$sex
pca_data$island <- penguins_clean$island
```

```r
pca_data|>
    ggplot(aes(x = PC1, y = PC2, color = species))+
    geom_point()+
    labs(
        title = "PCA Plot of species",
        x = "PC1",
        y = "PC2"
    )
pca_data|>
    ggplot(aes(x = PC1, y = PC2, color = sex))+
    geom_point()+
    labs(
        title = "PCA Plot of sex",
        x = "PC1",
        y = "PC2"
    )
pca_data|>
    ggplot(aes(x = PC1, y = PC2, color = island))+
    geom_point()+
    labs(
        title = "PCA Plot of island",
        x = "PC1",
        y = "PC2"
    )
```
